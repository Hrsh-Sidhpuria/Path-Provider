```python
from pickle import FALSE
from ipywidgets import interact, widgets


#Algorithm to find the shortest path betwwen two place.
def dijkstra(graph, start, end):
    distances = {node: float('infinity') for node in graph}
    predecessors = {node: None for node in graph}
    distances[start] = 0

    priority_queue = [(0, start)]

    while priority_queue:

      # this below two line is for viewing the path
        # print(f"Distances: {distances}")
        # print(f"Predecessors: {predecessors}")

        current_distance, current_node = min(priority_queue, key=lambda x: x[0])
        priority_queue.remove((current_distance, current_node))

        for neighbor, distance in graph[current_node]['distance'].items():
            total_distance = current_distance + distance
            if total_distance < distances[neighbor]:
                distances[neighbor] = total_distance
                predecessors[neighbor] = current_node
                priority_queue.append((total_distance, neighbor))

    path = []
    while end is not None:
        path.insert(0, end)
        end = predecessors[end]

    return path

# Create a list of coordinates for the polygon
coordinates = [(21.423686, 72.865880), (21.423584, 72.866237), (21.422909, 72.866214), (21.422929, 72.865787)]
coordinates_ENG = [(21.422864, 72.865956), (21.422855, 72.866235), (21.422362, 72.866208), (21.422365, 72.865932)]
coordinates_G2 = [(21.422840, 72.865936), (21.422838, 72.865768), (21.422368, 72.865739), (21.422368, 72.865907)]
coordinates_ground= [(21.423010, 72.865650), (21.422975, 72.864554), (21.422199, 72.864576), (21.422230, 72.865289),
                     (21.422457, 72.865286), (21.422460, 72.865629)]
coordinates_mess = [( 21.422346, 72.865746), (21.422333, 72.865908), (21.422212, 72.865906), (21.422225, 72.865760)]
coordinates_canteen = [(21.422421, 72.865298), (21.422424, 72.865652), (21.422290, 72.865646), (21.422305, 72.865310)]
coordinates_parking = [(21.423504, 72.866396), (21.423442, 72.866678), (21.422936, 72.866651), (21.422934, 72.866348)]

coordinates_science = [(21.422843, 72.866366), (21.422820, 72.866676), (21.422581, 72.866662), (21.422584, 72.866362)]
coordinates_nursing= [(21.422584, 72.866362 ), (21.422581, 72.866662), (21.422345, 72.866649), (21.422339, 72.866393),
 (21.422400, 72.866397), (21.422377, 72.866356)]

coordinates_workshop = [(21.422362, 72.866208), (21.422365, 72.865932), (21.422195, 72.865917), (21.422191, 72.866183)]
coordinates_h_parking = [(21.422190, 72.865089), (21.422232, 72.865642), (21.422156, 72.865642), (21.422146, 72.866161) ,
                        (21.421761, 72.866210), (21.421761, 72.865688), (21.421418, 72.865636),(21.421414, 72.865172)]

coordinates_hom_hospital= [(21.421741977451898, 72.86573139348225), (21.421741977451898, 72.86595140003394),
                         (21.421664291677406, 72.86595329664213), (21.421664291677406, 72.86599502202264), (21.421531872648345, 72.86600071184725),
                         (21.421519513532832, 72.86595329664213), (21.421440062051058, 72.86595329664213), (21.421427702927787, 72.86572191044122)]
coordinates_A_hospital= [(21.42174021186657, 72.86622071839896), (21.421738446281214, 72.86599502202264), (21.42141534380346, 72.86600450506366), (21.421410047035575, 72.86621502857435)]
coordinates_maingate = [(21.423790, 72.866247), (21.423723, 72.866219), (21.423657, 72.866379), (21.423724, 72.866401) ]
coordinates_school = [(21.423048, 72.866906), (21.423054, 72.867079), (21.422747, 72.867091), (21.422750, 72.866929) ]
coordinates_playground = [(21.423461, 72.866936), (21.423360, 72.867184), (21.423294, 72.867164), (21.423304, 72.867101),(21.423195, 72.867071), (21.423079, 72.867092), (21.423085, 72.866885),(21.423297, 72.866871) ]
coordinates_pharmacy = [(21.422292, 72.866365), (21.422295, 72.866661), (21.421999, 72.866659), (21.421993, 72.866378)]
coordinates_phy = [(21.421993, 72.866378), (21.421999, 72.866659), (21.421690, 72.866640), (21.421681, 72.866596)  , (21.421613, 72.866607), (21.421608, 72.866382)]

coordinates_unk = [(21.42326569513903, 72.86670333335935), (21.423417130265307, 72.86671449731664), (21.423504725021566, 72.86674320463534),
                   (21.423478001203183, 72.86680939095352)  , (21.423400056705002, 72.8667655325499), (21.423250106220063, 72.86677988620927)]
coordinates_unk1 = [(21.42176996354016, 72.86463086626274), (21.421768197955164, 72.86480725082575), (21.421656966057974, 72.86481483725858),
                   (21.421655200471616, 72.86492294392622)  , (21.421782322634463, 72.86492294392622), (21.42178055704963, 72.86509743188101),
                   (21.421498063200723, 72.86510312170563) ,(21.42146451701988, 72.86462896965455) ]
# Create a path coordinates

path1 = folium.PolyLine(((21.423714, 72.866327), (21.423628, 72.866303), (21.423239, 72.866290), (21.422862, 72.866272),  (21.422872, 72.865709), (21.422430, 72.865684), (21.422185, 72.865702), (21.422147, 72.866283),
               (21.421395, 72.866290), (21.421375, 72.865703), (21.421373, 72.865154)),color="#686e6b" , weight=15,opacity = 0.4, tooltip="path")
path2 = folium.PolyLine(((21.422883, 72.866275),(21.422776, 72.866291),(21.422340, 72.866275),(21.422134, 72.866273)),color="#686e6b" , weight=15,opacity = 0.4, tooltip="path")
path3 = folium.PolyLine(((21.422869, 72.866274),(21.422852, 72.866891),(21.423303, 72.866846),(21.423507, 72.866899)),color="#686e6b" , weight=15,opacity = 0.4, tooltip="path")
path4 = folium.PolyLine(((21.422856, 72.865796),(21.422871, 72.865714),(21.423020, 72.865717)),color="#686e6b" , weight=15,opacity = 0.4, tooltip="path")
path5 = folium.PolyLine(((21.422295, 72.866268),(21.422319, 72.866650)),color="#686e6b" , weight=8,opacity = 0.4, tooltip="path")
path6 = folium.PolyLine(((21.422247, 72.865695),(21.422250, 72.865256)),color="#686e6b" , weight=8,opacity = 0.4, tooltip="path")



# Create a polygon object
polygon_Recreation_Area = folium.Polygon(coordinates,width=0.1,color = "#c2582b",fill_color="green", tooltip="Recreation Area",opacity=0.8)
polygon_ENG = folium.Polygon(coordinates_ENG,width=0.1,color = "#c2582b",fill_color="#2bc2b0", tooltip="Degree/Diploma Engineering Collage",opacity=0.8)
polygon_G2 = folium.Polygon(coordinates_G2,width=0.1,color = "#c2582b",fill_color="#54de54", tooltip="Garden",opacity=0.8)
polygon_ground = folium.Polygon(coordinates_ground,width=0.1,color = "#b0853a",fill_color="#2e1003" , fill_opacity = 0.6, tooltip="Play Ground",opacity=0.8)
polygon_canteen = folium.Polygon(coordinates_mess,width=0.1,color = "#c2582b",fill_color="#efff08", tooltip="Canteen",opacity=0.8)

polygon_mess = folium.Polygon(coordinates_canteen,width=0.1,color = "#c2582b",fill_color="#fff308", tooltip="Mess",opacity=0.8)
polygon_parking = folium.Polygon(coordinates_parking,width=0.1,color = "#c2582b",fill_color="#08b5ff", tooltip="Parking",opacity=0.8)

polygon_science = folium.Polygon(coordinates_science,width=0.1,color = "#c2582b",fill_color="#b877d1", tooltip="Science Collage",opacity=0.8)
polygon_nursing = folium.Polygon(coordinates_nursing,width=0.1,color = "#c2582b",fill_color="#723a87", tooltip="Nursing Collage",opacity=0.8)
polygon_workshop = folium.Polygon(coordinates_workshop,width=0.1,color = "#c2582b",fill_color="#3f423d", tooltip="Work Shop",opacity=0.8)
polygon_h_parking = folium.Polygon(coordinates_h_parking,width=0.1,color = "#c2582b",fill_color="#634208",tooltip="Heavy Vehical Parking",opacity=0.8)
polygon_A_hospital = folium.Polygon(coordinates_A_hospital,width=0.1,color = "#c2582b",fill_color="#b877d1", tooltip="Proposed Ayurvedic Hospital",opacity=0.8)
polygon_hom_hospital= folium.Polygon(coordinates_hom_hospital,width=0.1,color = "#c2582b",fill_color="#b877d1", tooltip="Proposed Ayurvedic Collage",opacity=0.8)

polygon_maingate = folium.Polygon(coordinates_maingate,width=0.1,color = "#c2582b",fill_color="#f70000", tooltip="Vidhyadeep University Main Gate",opacity=0.8)
polygon_school = folium.Polygon(coordinates_school,width=0.1,color = "#c2582b",fill_color="#9077ed", tooltip="Vidhyadeep School",opacity=0.8)
polygon_playground = folium.Polygon(coordinates_playground,width=0.1,color = "#c2582b",fill_color="#00ed4b", tooltip="Play Ground",opacity=0.8)
polygon_pharmacy = folium.Polygon(coordinates_pharmacy,width=0.1,color = "#c2582b",fill_color="#d6d487", tooltip="Vidhyadeep Institute of Pharmacy(VIP)",opacity=0.8)
polygon_phy = folium.Polygon(coordinates_phy,width=0.1,color = "#c2582b",fill_color="#d6d487", tooltip="Physiotherapy",opacity=0.8)

polygon_unk = folium.Polygon(coordinates_unk,width=0.1,color = "#c2582b",fill_color="#6e5246", tooltip="unknown",opacity=0.8)
polygon_unk2 = folium.Polygon(coordinates_unk1,width=0.1,color = "#c2582b",fill_color="#dde366", tooltip="unknown1",opacity=0.8)


# giving marker and names

# map coordinates
map = folium.Map(location=[21.422452948205656, 72.86634182661665], zoom_start=20)


popup = folium.Popup(IFrame(html_content, width=250, height=300), max_width=500)
marker = folium.Marker(location=[21.423793, 72.866333], popup=popup, icon=folium.Icon(color='red'))
marker.add_to(map)

# ADDING CHILD TO MAP
map.add_child(polygon_Recreation_Area)
map.add_child(polygon_ENG)
map.add_child(polygon_G2)
map.add_child(polygon_ground)
map.add_child(polygon_mess)
map.add_child(polygon_canteen)
map.add_child(polygon_parking)

map.add_child(polygon_science)
map.add_child(polygon_nursing)
map.add_child(polygon_workshop)
map.add_child(polygon_h_parking)
map.add_child(polygon_A_hospital)

map.add_child(polygon_hom_hospital)
map.add_child(polygon_maingate)
map.add_child(polygon_school)
map.add_child(polygon_playground)
map.add_child(polygon_pharmacy)
map.add_child(polygon_phy)
map.add_child(polygon_unk)
map.add_child(polygon_unk2)



# ADDING PATH TO MAP
map.add_child(path1)
map.add_child(path2)
map.add_child(path3)
map.add_child(path4)
map.add_child(path5)
map.add_child(path6)




popup1 = folium.Popup(popup_eng)
popup2 = folium.Popup(popup_Recreation_Area)
popup3 = folium.Popup(popup_G2)
popup4 = folium.Popup(popup_ground)
popup5 = folium.Popup(popup_mess)
popup6 = folium.Popup(popup_canteen)
popup7 = folium.Popup(popup_parking)
popup8 = folium.Popup(popup_science)
popup9 = folium.Popup(popup_nursing)
popup10 = folium.Popup(popup_workshop)
popup11 = folium.Popup(popup_h_parking)
popup12 = folium.Popup(popup_A_hospital)
popup13 = folium.Popup(popup_hom_hospital)
popup14 = folium.Popup(popup_maingate)
popup15 = folium.Popup(popup_school)
popup16 = folium.Popup(popup_playground)
popup17 = folium.Popup(popup_pharmacy)
popup18 = folium.Popup(popup_phy)
popup19 = folium.Popup(popup_unk)
popup20 = folium.Popup(popup_unk2)


popup1.add_to(polygon_ENG)
popup2.add_to(polygon_Recreation_Area)
popup3.add_to(polygon_G2)
popup4.add_to(polygon_ground)
popup5.add_to(polygon_mess)
popup6.add_to(polygon_canteen)
popup7.add_to(polygon_parking)
popup8.add_to(polygon_science)
popup9.add_to(polygon_nursing)
popup10.add_to(polygon_workshop)
popup11.add_to(polygon_h_parking)
popup12.add_to(polygon_A_hospital)
popup13.add_to(polygon_hom_hospital)
popup14.add_to(polygon_maingate)
popup15.add_to(polygon_school)
popup16.add_to(polygon_playground)
popup17.add_to(polygon_pharmacy)
popup18.add_to(polygon_phy)
popup19.add_to(polygon_unk)
popup20.add_to(polygon_unk2)


#ADDING TEXT TO THE LOCATION

folium.Marker(location=[21.42335988928804, 72.86589851021863],
              popup = folium.Popup(IFrame(popup_Recreation_Area, width=250, height=300), max_width=500),
              tooltip='Recreation Area',
              icon=folium.DivIcon( html="""<div style="font-size: 12px;">Recreation<br>   Area</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.422673310826767, 72.86596391855792],
              popup = folium.Popup(IFrame(popup_eng, width=250, height=300), max_width=500),
              tooltip='Degree/Diploma Engineering College',
              icon=folium.DivIcon(html="""<div style="
               font-size: 7px;">Degree/Diploma <br> Engineering College</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.422612421453497, 72.86489483054343],
              popup = folium.Popup(IFrame(popup_playground, width=250, height=300), max_width=500),
              tooltip='PlayGround',
              icon=folium.DivIcon(html="""<div style="
               font-size: 14px;">Play Ground</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.42273000091073, 72.86641049964308],
              popup = folium.Popup(IFrame(popup_science, width=250, height=300), max_width=500),
              tooltip='Science College',
              icon=folium.DivIcon(html="""<div style="
               font-size: 9px;">Science College</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.422486443346394, 72.86638343412254],
              popup = folium.Popup(IFrame(popup_nursing, width=250, height=300), max_width=500),
              tooltip='Nursing College',
              icon=folium.DivIcon(html="""<div style="
               font-size: 9px;">Nursing College</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.422198792938673, 72.86638794503911],
              popup = folium.Popup(IFrame(popup_pharmacy, width=250, height=300), max_width=500),
              tooltip='Vidhyadeep Institute of Pharmacy(VIP)',
              icon=folium.DivIcon(html="""<div style="
               font-size: 9px;">Vidhyadeep Institute<br> of Pharmacy(VIP)</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.423240210705597, 72.86642177693508],
              popup = folium.Popup(IFrame(popup_parking, width=250, height=300), max_width=500),
              tooltip='Parking',
              icon=folium.DivIcon(html="""<div style="font-size: 12px;">Parking</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.42165722933222, 72.86585467301552],
              popup = folium.Popup(IFrame(popup_hom_hospital, width=250, height=150), max_width=500),
              tooltip='Vidhyadeep Homeopathic Hospital',
              icon=folium.DivIcon(html="""<div style="font-size: 10px;">Homeopathic <br>Hospital</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.42173491511047, 72.86638382670452],
              popup = folium.Popup(IFrame(popup_phy, width=250, height=300), max_width=500),
              tooltip='Vidhyadeep Institute Of Physiotherapy',
              icon=folium.DivIcon(html="""<div style="font-size: 10px;">Physiotherapy Hospital <br> and college</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.421899114465383, 72.86537483114158],
              popup = folium.Popup(IFrame(popup_h_parking, width=250, height=300), max_width=500),
              tooltip='Heavy Vehical Parking',
              icon=folium.DivIcon(html="""<div style="font-size: 12px;">Heavy Vehical Parking</div>""", class_name="mapText")).add_to(map)

folium.Marker(location=[21.422376540871614, 72.86542572701963],
              popup = folium.Popup(IFrame(popup_mess, width=250, height=300), max_width=500),
              tooltip='Vidhyadeep Mess',
              icon=folium.DivIcon(html="""<div style="font-size: 10px;">Mess</div>""", class_name="mapText")).add_to(map)

#STYLE SHEET DOR LOCATION TEXT
map.get_root().html.add_child(folium.Element("""
<style>
.mapText {
    white-space: nowrap;
    color: white;
    text-align: center;
    text-shadow: -1px -1px 1px black, 1px -1px 1px black, -1px 1px 1px black,
                 1px 1px 1px black;
}
</style>
"""))



coordinates = {
    'Gate': {'coords': (21.42379034150547, 72.8663158396076), 'distance': {'cross A': 101, 'cross B': 183}},
    'cross A': {'coords': (21.422835716322737, 72.86626681815257), 'distance': {'Gate': 101, 'cross B': 80, 'turn 1': 61 , 'turn E':30,'turn G1':45}},
    'cross B': {'coords': (21.42213706509904, 72.86624341788765), 'distance': {'Gate': 180, 'cross A': 80,'turn 2': 61}},
    'turn 1': {'coords': (21.422858613112137, 72.86569478912917), 'distance': {'cross A': 61,'turn G2':52, 'turn E': 44, 'turn 2': 80,'turn C' : 65 ,'turn M' :55,'turn G1':17 }},
    'turn 2': {'coords': (21.422199421248195, 72.86569478912917), 'distance': {'turn G2':25 ,'cross B': 61, 'turn C': 7, 'turn 1': 80}},
    'turn E': {'coords': (21.422860979971137, 72.8660643201342), 'distance': {'turn 1': 44, 'cross A': 17,'Eng':10 ,'turn G1':20}},
    'turn C': {'coords': (21.422260728910572, 72.86567642656198), 'distance': {'turn 2': 7,'turn G2':8, 'Canteen': 7, 'turn M': 10,'turn 1': 65}},
    'Canteen': {'coords': (21.42224291288044, 72.86581039404935), 'distance': {'turn C':7, 'turn M': 20, 'Mess': 25}},
    'turn M': {'coords': (21.422352778365013, 72.86567642656198), 'distance': {'turn G2':10,'turn C': 10,'turn 1': 55, 'Mess': 7}},
    'Mess': {'coords': (21.422355747701285, 72.86544676801219), 'distance': {'turn M': 7 }},
    'turn G1': {'coords': (21.422858675743413, 72.86582235314415), 'distance': {'Garden': 40,'turn E':20,'cross A':45,'turn 1':17 }},
    'turn G2': {'coords': (21.42237515445941, 72.8656909554272), 'distance': {'Garden': 40,'turn 1':52,'turn C':8,'turn M':10,'turn 2':25 }},
    'Garden': {'coords': (21.422684859794835, 72.86583503315714), 'distance': {'turn G2': 40,'turn M': 40 }},
    'Eng': {'coords': (21.422678406654967, 72.86607148188315), 'distance': {'turn E': 10}}


}
print(" Please select the place : \n")

# Define the update function to display the shortest path
def update_map(from_location, to_location):
    # Clear previous map
    output.clear_output(wait=True)
    # Calculate shortest path
    shortest_path = dijkstra(coordinates, from_location, to_location)
    ant_path_coords = [coordinates[location]['coords'] for location in shortest_path]
    # Add the shortest path
    plugins.AntPath(locations=ant_path_coords, dash_array=[10, 20], delay=800, color='blue').add_to(map)
    # Display the map
    with output:
        display(map)




# Define the options for the dropdown menus
from_options = ['Gate', 'Canteen','Mess', 'Garden', 'Eng']
to_options = ['Gate', 'Canteen','Mess', 'Garden', 'Eng']

# Create dropdown menus for selecting "from" and "to" locations with selective options
from_dropdown = widgets.Dropdown(options=from_options, description='From:')
to_dropdown = widgets.Dropdown(options=to_options, description='To:')

# Define a function to handle button click event
def on_submit_button_clicked(b):
    update_map(from_dropdown.value, to_dropdown.value)

# Create a submit button
submit_button = Button(description="Find Path")
submit_button.on_click(on_submit_button_clicked)

# Create an output widget to display the map
output = Output()

# Display the dropdown menus and the submit button
display(from_dropdown, to_dropdown, submit_button, output)
# # Add markers for each location
# for location, data in coordinates.items():
#     folium.Marker(location=data['coords'], popup=location).add_to(map)


# # Create dropdown menus for selecting "from" and "to" locations
# from_dropdown = widgets.Dropdown(options=list(coordinates.keys()), description='From:')
# to_dropdown = widgets.Dropdown(options=list(coordinates.keys()), description='To:')

# # Define a function to update the map based on selected locations
# def update_map(from_location, to_location):
#     shortest_path = dijkstra(coordinates, from_location, to_location)
#     ant_path_coords = [coordinates[location]['coords'] for location in shortest_path]
#     map = folium.Map(location=[21.422452948205656, 72.86634182661665], zoom_start=20)
#     plugins.AntPath(locations=ant_path_coords, dash_array=[10, 20], delay=800, color='blue').add_to(map)
#     display(map)

# # Use the interact function to connect the dropdown menus with the update_map function
# interact(update_map, from_location=from_dropdown, to_location=to_dropdown)
```
